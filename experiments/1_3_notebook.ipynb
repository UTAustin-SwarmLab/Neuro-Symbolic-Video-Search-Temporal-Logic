```python
threshold_data = []
precision_data = []
accuracy_data = []
recall_data = []
for threshold in data.keys():
    for result in data[threshold]:
        print(result)
        threshold_data.append(float(threshold))
        accuracy_data.append(result['accuracy'])
        precision_data.append(result['precision'])
        recall_data.append(result['recall'])


```

```python
import seaborn as sns


ax = sns.lineplot(x=threshold_data, y=accuracy_data, label='accuracy', linestyle='dashed')
sns.lineplot(x=threshold_data, y=precision_data, ax=ax, label='precision', linestyle='dashed')
sns.lineplot(x=threshold_data, y=recall_data, label='racall', linestyle='dashed')

ax.set_xlabel('confidence')
ax.set_ylabel('accuracy')
```
